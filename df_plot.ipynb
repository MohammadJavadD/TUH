```python
from numpy import genfromtxt
import matplotlib.pyplot as plt
```

```python
my_data = genfromtxt('accuracy_runs.csv', delimiter=',')
```

```python
plt.plot(my_data[:,5],my_data[:,2], linestyle = 'dotted', label='train acc')
plt.plot(my_data[:,5],my_data[:,3], linestyle = 'dotted', label='val acc')
plt.plot(my_data[:,5],my_data[:,4], linestyle = 'dotted', label='test acc')


plt.legend(title='Metric:')
plt.xlabel('number of sessions in training ')
plt.show()
```
